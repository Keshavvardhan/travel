```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one=OneHotEncoder(sparse=False,handle_unknown='ignore',drop='first')
```

```python
x_dup= pd.DataFrame(one.fit_transform(x.iloc[:,0:4]),columns=one.get_feature_names_out())
```

```python
x_dup
```

```python
x.drop(columns=["Car_Name","Fuel_Type","Seller_Type","Transmission"],inplace=True)
```

```python
x
```

```python
x=pd.concat([x,x_dup],axis=1)
```

```python
x
```

```python
import joblib
```

```python
joblib.dump(one,"OneHotecar.joblib")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
       "n_estimators":[100,500],
       "max_features":['auto','sqrt','log2'],
       "max_depth":[4,5,6,7,8],
        "criterion":['squared_error']
}
```

```python
grid_model = GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
regressor=RandomForestRegressor(n_estimators=100,max_features='auto',max_depth=8,criterion='squared_error')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=32)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
pre=regressor.predict(x_test)
```

```python
pre
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,pre)
```

```python
r2_score(y_test,pre)
```

```python
joblib.dump(regressor,('regressorcar.pkl'))
```
